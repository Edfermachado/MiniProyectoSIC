```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


df = pd.read_csv('data_science_salaries.csv')


x = df[['work_year']]  
y = df['salary_in_usd']


model = LinearRegression()


model.fit(x, y)


print("Pendiente (m):", model.coef_[0])
print("Interseccin (b):", model.intercept_)


y_pred = model.predict(x)

plt.scatter(x, y, color='blue', label='Datos originales')
plt.plot(x, y_pred, color='red', label='Recta de regresin')
plt.xlabel('Ao de trabajo (work_year)')
plt.ylabel('Salario en USD (salary_in_usd)')
plt.legend()
plt.show()
```

```python
import pandas as pd


df = pd.read_csv('data_science_salaries.csv')

x = df['work_year']
y = df['salary_in_usd']


n = len(x)
sum_x = x.sum()
sum_y = y.sum()
sum_xy = (x * y).sum()
sum_x_squared = (x ** 2).sum()


m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
b = (sum_y - m * sum_x) / n


year_2025 = 2024
salary_2025 = m * year_2025 + b

print("Pendiente (m):", m)
print("Interseccin (b):", b)
print("Salario estimado en 2025:", salary_2025)


m_formatted = f"{m:,.2f}"
b_formatted = f"{b:,.2f}"
salary_2025_formatted = f"{salary_2025:,.2f}"

print("#########################################")

print("Pendiente (m):", m_formatted)
print("Interseccin (b):", b_formatted)
print("Salario estimado en 2025: $", salary_2025_formatted)

```

```python
import pandas as pd

df = pd.read_csv('data_science_salaries.csv')

df_data_engineer = df[df['job_title'] == 'Data Engineer']

x = df_data_engineer['work_year']
y = df_data_engineer['salary_in_usd']

n = len(x)
sum_x = x.sum()
sum_y = y.sum()
sum_xy = (x * y).sum()
sum_x_squared = (x ** 2).sum()

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
b = (sum_y - m * sum_x) / n

year_2025 = 2025
salary_2025 = m * year_2025 + b

m_formatted = f"{m:,.2f}"
b_formatted = f"{b:,.2f}"
salary_2025_formatted = f"{salary_2025:,.2f}"

print("Pendiente (m):", m_formatted)
print("Interseccin (b):", b_formatted)
print("Salario estimado en 2025:", salary_2025_formatted)
```
